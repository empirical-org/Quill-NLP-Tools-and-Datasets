```python
import spacy

nlp = spacy.load("/tmp/grammar_vfg")
#nlp = spacy.load("../models/grammar_perfect_progressive_with_incorrect_be_and_without_have_spacy")
```

```python
def find_errors(sentences):
    total = 0
    for sentence in sentences:
        doc = nlp(sentence)
        
        print(doc)
        print(doc.ents)
        for entity in doc.ents:
            #if entity.label_.startswith("Subject_verb agreement"): 
            #if entity.label_.startswith("Past"):
            if entity.label_.startswith("Passive"):
                total += 1
                break
        
    return total

true_positives = find_errors(incorrect_sentences)
false_positives = find_errors(correct_sentences)

print("TP:", true_positives)
print("FP:", false_positives)

p = true_positives/(true_positives+false_positives) if (true_positives+false_positives) > 0 else 0
r = true_positives/len(incorrect_sentences)
f = 2*p*r/(p+r) if (p+r) > 0 else 0

print("P:", p)
print("R:", r)
print("F:", f)
print("A:", (true_positives + (100-false_positives)) / 200)
```
